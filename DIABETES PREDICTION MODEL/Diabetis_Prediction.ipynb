IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

ANALYSIS OF THE DATA

```python
diabetes_data.shape
```

```python
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
# NON DIABETIC = 0
# DIABETIC = 1
```

ANALYSIS OF DIABETIC AND NON INDIVIDUALS

```python
diabetes_data.groupby('Outcome').mean()
#The glucose level is significantly high in diabetic individual, so is the insulin level.
```

```python
A = diabetes_data.drop(columns='Outcome', axis=1)
B = diabetes_data['Outcome']
print(A, B)
```

DATA STANDADIZATION

```python
scaler = StandardScaler()
scaler.fit(A)
standardized_data = scaler.transform(A)
print(standardized_data)

```

TRAIN AND TEST DATA SPLITTING

```python
A_test, A_train, B_test, B_train = train_test_split(A, B, test_size = 0.2, stratify = B, random_state = 2)
print(A.shape, A_train.shape, A_test.shape)
```

TRAINING OUR MODEL


```python
classified = svm.SVC(kernel='linear')
classified.fit(A_train, B_train)
```

DETERMINE ACCURACY ON TRAINING DATA

```python
A_train_prediction = classified.predict(A_train)
training_data_accuracy = accuracy_score(A_train_prediction, B_train)
print('Accuracy score of the training data : ', training_data_accuracy)

A_test_prediction = classified.predict(A_test)
test_data_accuracy = accuracy_score(A_test_prediction, B_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

MAKING A PREDICTION

```python
medical_data = (6,148,72,35,0,33.6,0.627,50)
#changing the medical data to numpy array
medical_data_as_numpy_array = np.asarray(medical_data)

#reshape the array
medical_data_reshaped = medical_data_as_numpy_array.reshape(1,-1)

#Standadizing the medical_data
standard_data = scaler.transform(medical_data_reshaped)
print(standard_data)

prediction = classified.predict(standard_data)
print(prediction)
if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
